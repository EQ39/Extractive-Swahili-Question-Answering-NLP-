```python
from transformers import AutoTokenizer

# Load the JSON data
with open('train.json', 'r') as file:
    data = json.load(file)

# Access the train data
train_data = data['data']

# Initialize the tokenizer
trained_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(trained_checkpoint)

def train_data_preprocess(examples):
    """
    Generate start and end indexes of answer in context
    """

    def find_context_start_end_index(sequence_ids):
        """
        Returns the token index in which context starts and ends
        """
        token_idx = 0
        while sequence_ids[token_idx] != 1:  # means its special tokens or tokens of question
            token_idx += 1                   # loop only breaks when context starts in tokens
        context_start_idx = token_idx

        while sequence_ids[token_idx] == 1:
            token_idx += 1
        context_end_idx = token_idx - 1
        return context_start_idx, context_end_idx

    questions = []
    contexts = []
    answers = []

    for example in examples:
        for paragraph in example['paragraphs']:
            context = paragraph['context']
            for qa in paragraph['qas']:
                question = qa['question'].strip()
                answer = qa['answers'][0]

                questions.append(question)
                contexts.append(context)
                answers.append(answer)

    inputs = tokenizer(
        questions,
        contexts,
        max_length=512,
        truncation="only_second",
        stride=128,
        return_overflowing_tokens=True,  # returns id of base context
        return_offsets_mapping=True,     # returns (start_index,end_index) of each token
        padding="max_length"
    )

    start_positions = []
    end_positions = []

    for i, mapping_idx_pairs in enumerate(inputs['offset_mapping']):
        context_idx = inputs['overflow_to_sample_mapping'][i]

        # from main context
        answer = answers[context_idx]
        answer_start_char_idx = answer['answer_start']
        answer_end_char_idx = answer_start_char_idx + len(answer['text'])

        # now we have to find it in sub contexts
        tokens = inputs['input_ids'][i]
        sequence_ids = inputs.sequence_ids(i)

        # finding the context start and end indexes wrt sub context tokens
        context_start_idx, context_end_idx = find_context_start_end_index(sequence_ids)

        # starting and end index of character of full context text
        context_start_char_index = mapping_idx_pairs[context_start_idx][0]
        context_end_char_index = mapping_idx_pairs[context_end_idx][1]

        # If the answer is not fully inside the context, label is (0, 0)
        if (context_start_char_index > answer_start_char_idx) or (
            context_end_char_index < answer_end_char_idx):
            start_positions.append(0)
            end_positions.append(0)

        else:
            # else its start and end token positions
            # here idx indicates index of token
            idx = context_start_idx
            while idx <= context_end_idx and mapping_idx_pairs[idx][0] <= answer_start_char_idx:
                idx += 1
            start_positions.append(idx - 1)

            idx = context_end_idx
            while idx >= context_start_idx and mapping_idx_pairs[idx][1] > answer_end_char_idx:
                idx -= 1
            end_positions.append(idx + 1)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs

# Select a subset of the training data
train_sample = dataset["train"][:200]
# Preprocess the training data
#train_dataset = train_data_preprocess(train_sample)

train_dataset = train_data_preprocess(train_sample)

print(f"Original train data length: {len(train_data)}")
print(f"Preprocessed train data length: {len(train_dataset['input_ids'])}")
```

```python
from transformers import AutoTokenizer

def preprocess_validation_examples(examples):
    """
    Preprocess validation data
    """
    questions = []
    contexts = []
    ids = []

    for example in examples:
        for paragraph in example['paragraphs']:
            context = paragraph['context']
            for qa in paragraph['qas']:
                question = qa['question'].strip()
                id_ = qa['id']

                questions.append(question)
                contexts.append(context)
                ids.append(id_)

    inputs = tokenizer(
        questions,
        contexts,
        max_length=512,
        truncation="only_second",
        stride=128,
        return_overflowing_tokens=True,
        return_offsets_mapping=True,
        padding="max_length",
    )

    sample_map = inputs.pop("overflow_to_sample_mapping")

    base_ids = []

    for i in range(len(inputs["input_ids"])):
        # take the base id (i.e., in cases of overflow, we get the base id)
        base_context_idx = sample_map[i]
        base_ids.append(ids[base_context_idx])

        # sequence id indicates the input: 0 for first input, 1 for second input,
        # and None for special tokens by default
        sequence_ids = inputs.sequence_ids(i)
        offset = inputs["offset_mapping"][i]
        # for question tokens, provide offset_mapping as None
        inputs["offset_mapping"][i] = [
            o if sequence_ids[k] == 1 else None for k, o in enumerate(offset)
        ]

    inputs["base_id"] = base_ids
    return inputs

trained_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(trained_checkpoint)

data_val_sample =validation_data = data['data'][:31]
eval_set = preprocess_validation_examples(data_val_sample)
print(len(eval_set['input_ids']))
```

```python
import numpy as np
import collections
import evaluate

def predict_answers_and_evaluate(start_logits, end_logits, eval_set, examples):
    example_to_features = collections.defaultdict(list)
    for idx, base_id in enumerate(eval_set["base_id"]):
        example_to_features[base_id].append(idx)

    n_best = 20
    max_answer_length = 30
    predicted_answers = []

    for example in examples:
        for paragraph in example["paragraphs"]:
            context = paragraph["context"]
            for qa in paragraph["qas"]:
                example_id = qa["id"]
                answers = []

                for feature_index in example_to_features[example_id]:
                    start_logit = start_logits[feature_index]
                    end_logit = end_logits[feature_index]
                    offsets = eval_set[feature_index].offsets

                    start_indexes = np.argsort(start_logit).tolist()[::-1][:n_best]
                    end_indexes = np.argsort(end_logit).tolist()[::-1][:n_best]

                    #print(f"Context: {context}")
                    #print(f"Offsets: {offsets}")
                    #print("Start Logits:", start_logit)
                    #print("End Logits:", end_logit)
                    #print("Start Indexes:", start_indexes)
                    #print("End Indexes:", end_indexes)

                    for start_index in start_indexes:
                        for end_index in end_indexes:
                            if offsets[start_index] is None or offsets[end_index] is None:
                                continue
                            if (
                                end_index < start_index
                                or end_index - start_index + 1 > max_answer_length
                            ):
                                continue

                            start_char = offsets[start_index][0]
                            end_char = offsets[end_index][1]
                            predicted_text = context[start_char:end_char]
                            logit_score = start_logit[start_index] + end_logit[end_index]
                            answers.append({
                                "text": predicted_text,
                                "logit_score": logit_score,
                            })

                            # Debug print for each answer
                            #print(f"Start Index: {start_index}, End Index: {end_index}")
                            #print(f"Predicted Text: {predicted_text}")
                            #print(f"Logit Score: {logit_score}")

                if len(answers) > 0:
                    best_answer = max(answers, key=lambda x: x["logit_score"])
                    predicted_answers.append(
                        {"id": example_id, "prediction_text": best_answer["text"]}
                    )
                else:
                    predicted_answers.append({"id": example_id, "prediction_text": ""})

    metric = evaluate.load("squad")

    theoretical_answers = []
    for example in examples:
        for paragraph in example["paragraphs"]:
            for qa in paragraph["qas"]:
                answer_starts = [ans["answer_start"] for ans in qa["answers"]]
                answer_texts = [ans["text"] for ans in qa["answers"]]
                answers = [{"text": text, "answer_start": start} for text, start in zip(answer_texts, answer_starts)]
                theoretical_answers.append({"id": str(qa["id"]), "answers": answers})

    print("Predicted Answers:")
    for pa in predicted_answers:
        print(pa)

    print("\nTheoretical Answers:")
    for ta in theoretical_answers:
        print(ta)

    # Convert predictions to the expected format
    formatted_predictions = [{'id': str(prediction['id']), 'prediction_text': prediction['prediction_text']} for prediction in predicted_answers]

    # Convert references to the expected format
    formatted_references = [{'id': reference['id'], 'answers': [{'text': ans['text'], 'answer_start': ans['answer_start']} for ans in reference['answers']]} for reference in theoretical_answers]

    # Compute the metric
    metric_ = metric.compute(predictions=formatted_predictions, references=formatted_references)


    return predicted_answers, metric_


pred_answers, metrics_ = predict_answers_and_evaluate(start_logits, end_logits, eval_set, data_val_sample)
print(metrics_)
```

```python
def error_analysis(predictions, references):
    # Initialize dictionaries to store error statistics
    error_types = {
        "empty_prediction": 0,
        "incorrect_prediction": 0,
        "partial_prediction": 0
    }

    detailed_errors = []

    # Iterate over predictions and references to analyze errors
    for pred, ref in zip(predictions, references):
        pred_text = pred["prediction_text"]
        ref_texts = [ans["text"] for ans in ref["answers"]]

        if not pred_text:
            error_types["empty_prediction"] += 1
            detailed_errors.append({
                "id": pred["id"],
                "error_type": "empty_prediction",
                "prediction": pred_text,
                "reference": ref_texts
            })
            continue

        if pred_text not in ref_texts:
            if any(pred_text in ref_text for ref_text in ref_texts):
                error_types["partial_prediction"] += 1
                error_type = "partial_prediction"
            else:
                error_types["incorrect_prediction"] += 1
                error_type = "incorrect_prediction"

            detailed_errors.append({
                "id": pred["id"],
                "error_type": error_type,
                "prediction": pred_text,
                "reference": ref_texts
            })

    # Print error analysis results
    print("Error Analysis Results:")
    print("-----------------------")
    total_errors = sum(error_types.values())
    print(f"Total Errors: {total_errors}")
    for error_type, count in error_types.items():
        print(f"{error_type.capitalize().replace('_', ' ')}: {count} ({(count / total_errors) * 100:.2f}%)")

    # Print detailed error cases
    print("\nDetailed Errors:")
    for error in detailed_errors:
        print(f"ID: {error['id']}")
        print(f"Error Type: {error['error_type']}")
        print(f"Prediction: {error['prediction']}")
predicted_answers = [
    {'id': 10189, 'prediction_text': 'Mwendo wa kobe humfikisha mba'},
    {'id': 10233, 'prediction_text': '\n Nilichag'},
    {'id': 10234, 'prediction_text': 'Mazingira safi'},
    {'id': 10190, 'prediction_text': 'Kutokana na jina lake chanjagaa mzee huyu alikuwa amependa kuchanja kuni ili'},
    {'id': 10191, 'prediction_text': 'Tulipokaribia kufika kanisani, tulisikia sauti za wanakway'},
    {'id': 10192, 'prediction_text': 'Humo mbing'},
    {'id': 10193, 'prediction_text': 'kwa upande mwingine wan'},
    {'id': 10194, 'prediction_text': 'Mzazi'},
    {'id': 10195, 'prediction_text': 'wazazi wangu wanape'},
    {'id': 10099, 'prediction_text': 'Wengi mungeuliza ya kwamba ajali ni nini? Kwa upand'},
    {'id': 10100, 'prediction_text': 'madereva wasiojali wengine wapokonywe leseni zao na shirika la uc'},
    {'id': 10101, 'prediction_text': 'mchelea mwana kulia hulia mw'},
    {'id': 10102, 'prediction_text': 'Mwend'},
    {'id': 10196, 'prediction_text': ' Mazao yanayotokana na kilimo zitasaf'},
    {'id': 10197, 'prediction_text': 'wapa Cheti (lisensi) inayodhihirisha kuwa biashara hiyo imekubalika na serikali bi'},
    {'id': 10198, 'prediction_text': 'Vi'},
    {'id': 10199, 'prediction_text': 'Serikali inapaswa kuweka uthibiti mwendo kwa kila barabara ili'},
    {'id': 10103, 'prediction_text': 'tayari kwa madereva na rangi'},
    {'id': 10104, 'prediction_text': 'husababishwa'},
    {'id': 10105, 'prediction_text': 'Matwana moja ilipita ikiyumba yumba kwa kusheheni mizigo na abiri'},
    {'id': 10106, 'prediction_text': 'Siasa hutokea mahali popote ambapo binadamu huchagua kiongozi'},
    {'id': 10107, 'prediction_text': 'Kwanza kabisa shu'},
    {'id': 10108, 'prediction_text': 'Ugonjwa wa korona ni ugonjwa unaoenezwa na virusi'},
    {'id': 10109, 'prediction_text': 'Kilimo huleta chakula cha kutosha kwa wan'},
    {'id': 10200, 'prediction_text': 'Basi wanyama hao walipofika maji walif'},
    {'id': 10201, 'prediction_text': 'Kuenea kwa mawasiliano na utandawazi kunaongeza utalii'},
    {'id': 10110, 'prediction_text': 'kiujum'},
    {'id': 10202, 'prediction_text': 'Mazungumzo'},
    {'id': 10235, 'prediction_text': 'wa wa tumbo wa kuhara na kutapika maji yote uliy'},
    {'id': 10111, 'prediction_text': 'Kilimo ni njia au sayansi'},
    {'id': 10203, 'prediction_text': 'kinachomtegemea mw'}
]
theoretical_answers = [
    {'id': '10189', 'answers': [{'text': ' mbali', 'answer_start': 25}]},
    {'id': '10233', 'answers': [{'text': 'Chapakazi', 'answer_start': 73}]},
    {'id': '10234', 'answers': [{'text': 'binadamu', 'answer_start': 48}]},
    {'id': '10190', 'answers': [{'text': 'kuni', 'answer_start': 68}]},
    {'id': '10191', 'answers': [{'text': 'sauti za wanakwaya', 'answer_start': 41}]},
    {'id': '10192', 'answers': [{'text': 'almasi ', 'answer_start': 109}]},
    {'id': '10193', 'answers': [{'text': 'moshi', 'answer_start': 138}]},
    {'id': '10194', 'answers': [{'text': 'hadithi', 'answer_start': 43}]},
    {'id': '10195', 'answers': [{'text': 'kilimo', 'answer_start': 63}]},
    {'id': '10099', 'answers': [{'text': 'ajali', 'answer_start': 77}]},
    {'id': '10100', 'answers': [{'text': 'madereva wasiojali wengine', 'answer_start': 0}]},
    {'id': '10101', 'answers': [{'text': 'hulia mwenyewe', 'answer_start': 35}]},
    {'id': '10102', 'answers': [{'text': 'Mwendokasi ', 'answer_start': 0}]},
    {'id': '10196', 'answers': [{'text': ' bidhaa', 'answer_start': 102}]},
    {'id': '10197', 'answers': [{'text': '(lisensi', 'answer_start': 141}]},
    {'id': '10198', 'answers': [{'text': 'tumia dawa za kulevya', 'answer_start': 70}]},
    {'id': '10199', 'answers': [{'text': ' uthibiti mwendo', 'answer_start': 24}]},
    {'id': '10103', 'answers': [{'text': 'tatu', 'answer_start': 11}]},
    {'id': '10104', 'answers': [{'text': 'mwendo kasi', 'answer_start': 49}]},
    {'id': '10105', 'answers': [{'text': 'moja ', 'answer_start': 8}]},
    {'id': '10106', 'answers': [{'text': 'wananchi ', 'answer_start': 75}]},
    {'id': '10107', 'answers': [{'text': 'maendeleo', 'answer_start': 41}]},
    {'id': '10108', 'answers': [{'text': 'virusi', 'answer_start': 43}]},
    {'id': '10109', 'answers': [{'text': 'Kilimo', 'answer_start': 0}]},
    {'id': '10200', 'answers': [{'text': ' maji', 'answer_start': 27}]},
    {'id': '10201', 'answers': [{'text': 'kikanda', 'answer_start': 58}]},
    {'id': '10110', 'answers': [{'text': 'mimea', 'answer_start': 27}]},
    {'id': '10202', 'answers': [{'text': 'gharama ', 'answer_start': 18}]},
    {'id': '10235', 'answers': [{'text': 'maji chafu', 'answer_start': 196}]},
    {'id': '10111', 'answers': [{'text': 'Kilimo', 'answer_start': 0}]},
    {'id': '10203', 'answers': [{'text': 'utegemezi', 'answer_start': 30}]}
]
error_analysis(predicted_answers, theoretical_answers)
```

```python
def preprocess_validation_examples(examples):
    questions = [q.strip() for q in examples["question"]]
    contexts = examples["context"]

    inputs = tokenizer(
        questions,
        contexts,
        max_length=512,
        truncation="only_second",
        stride=128,
        return_overflowing_tokens=True,
        return_offsets_mapping=True,
        padding="max_length",
    )

    sample_map = inputs.pop("overflow_to_sample_mapping")
    example_ids = []

    for i in range(len(inputs["input_ids"])):
        sample_idx = sample_map[i]
        example_ids.append(examples["id"][sample_idx])

        sequence_ids = inputs.sequence_ids(i)
        offset = inputs["offset_mapping"][i]
        inputs["offset_mapping"][i] = [
            o if sequence_ids[k] == 1 else None for k, o in enumerate(offset)
        ]

    inputs["example_id"] = example_ids
    return inputs
```

```python
# Postprocess the model's output
def postprocess_output(output, inputs):
    answer_start = torch.argmax(output.start_logits)
    answer_end = torch.argmax(output.end_logits) + 1
    input_ids = inputs["input_ids"][0]
    offsets = inputs["offset_mapping"][0]

    start_char = offsets[answer_start][0].item()
    end_char = offsets[answer_end][1].item()

    answer = tokenizer.decode(input_ids[answer_start:answer_end])

    return answer, start_char, end_char

answer, start_char, end_char = postprocess_output(outputs, inputs)

# Print the question, answer, and the corresponding span in the context
print("Question:", question)
print("Answer:", answer)
print("Context:")
print(context)
print(f"Answer span: ({start_char}, {end_char})")
print(context[start_char:end_char])
```

```python
# BERT only needs the token IDs, but for the purpose of inspecting the
# tokenizer's behavior, let's also get the token strings and display them.
tokens = tokenizer.convert_ids_to_tokens(input_ids)

# For each token and its id...
for token, id in zip(tokens, input_ids):

    # If this is the [SEP] token, add some space around it to make it stand out.
    if id == tokenizer.sep_token_id:
        print('')

    # Print the token string and its ID in two columns.
    print('{:<12} {:>6,}'.format(token, id))

    if id == tokenizer.sep_token_id:
        print('')
```

```python
# Search the input_ids for the first instance of the `[SEP]` token.
sep_index = input_ids.index(tokenizer.sep_token_id)

# The number of segment A tokens includes the [SEP] token istelf.
num_seg_a = sep_index + 1

# The remainder are segment B.
num_seg_b = len(input_ids) - num_seg_a

# Construct the list of 0s and 1s.
segment_ids = [0]*num_seg_a + [1]*num_seg_b

# There should be a segment_id for every input token.
assert len(segment_ids) == len(input_ids)
```
